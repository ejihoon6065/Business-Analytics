# 1. 소셜 네트워크 이론의 개요 및 역사
## 1.1. 소셜 네트워크 이론의 개요
- 네트워크: 각각의 객체들이 상호 연결된 구조
- 구성요소: node와 링크
- 노드: 고유한 속성을 가지고 있는 행위자
- 링크: 노드들간의 관계
   
   => 네트워크는 복수의 노드들이 연결된 구조

## 1.2. 네트워크 이론의 역사
1. 버트의 "구조적 공백"
    - 사회 네트워크 구조 내에서 상대적으로 중요한 위치와 역할을 하는 노드들을 설명하기 위해 ‘구조적 공백(Structural holes)’ 이론을 제시
    - 구조적 공백의 위치에 있는 노드는 다른 노드들과 중복적인 관계를 가지고 있지 않기 때문에 노드와 노드를 연결시켜 주는 역할을 함
    - 버트의 구조적 공백은 그라노베티의 ‘약한 연결(weak ties)’ 개념을 더 구체화 한 것으로 볼 수 있음
    
2. 바라바시, 알버트, 정하웅의 "무척도 네트워크"
    - 하이퍼 링크로 연결된 인터넷의 물리적 연결 구조에 대한 논문에서 인터넷 연결 수 분포가 정규분포를 따르지 않고 멱함수법칙(Power law)을따르는 것을 밝혔으며, 이러한 네트워크 구조
    - 소수의 링크만을 가지고 있는 수많은 노드들과 수많은 링크를 가지고 있는 소수의 노드들로 양분되는 것.

# 2. 네트워크 이론의 기초
## 2.1. 그래프의 기본 개념
- 그래프 이론(Graph theory) – 객체들 간의 네트워크 관계 모델을 수학적으로 표현된 그래프로 나타내는연구 분야 그래프는 노드(Node, vertex)로 구성, 이러한 노드들은 링크(Link, edge)로 연결되어 있음.
- 무방향(Undirected) 네트워크: 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크 등
- 방향(Directed) 네트워크: 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등
- 네트워크 그래프에서 노드와 노드 간 관계의 유무만이 아니라 관계의 정도(강도)를 함께 표시할 수 있음
- 관계의 유무만을 표현한 경우를 ‘이진 네트워크(binary network)’ 그래프라고 하고, 관계의 정도를 함께 표시한 경우를 ‘가중 네트워크(valued network)’ 그래프라 함.

## 2.2. 그래프의 표현 방법
- 인접행렬
    - 인접한 노드들 간의 연결 관계를 행렬(Matrix)로 표현, 네트워크 상에서 노드의 개수가 n인 경우 인접 행렬은 nxn의 정방 행렬로 표현
- 노드 리스트
    - 인접한 링크가 있는 노들만을 표기하는 방법으로 노드들만의 목록
- 엣지 리스트
    - 노드들을 연결한 링크들의 리스트로 “링크의 목록＂

# 3. 소셜 네트워크 분석 지표
## 3.1. 네트워크 수준의 속성 분석
- 네트워크 크기: 네트워크를 구성하는 노드들의 수
- 네트워크 밀도: 네트워크를 구성하는 노드들 간의 연결된 정도
- 포괄성: 네트워크상의 총 노드의 개수 중에서 실제로 연결 관계가 존재하는 노드의 개수가 차지하는 비율
- 연결정도: 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수
- 연결거리: 노드간 관계(링크)의 강도
- 직경
- 평균연결거리
- 상호성: 상호성 계산은 전체 네트워크 상에서 상호적 연결이 차지하고 있는 비율로 산출
- 군집화 계수:군집화의 정도를 나타내는 지표를 뜻하며 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중에서 실제로 존재하는 관계의 수가 차지하는 비율로 정의


## 3.2. 중심성 분석
- 중심성 지표
    - 중심성 지표는 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표
    - 연결정도 중심성
    - 근접 중심성
    - 매개 중심성
    - 아이겐벡터 중심성

# 4. 소셜 네트워크 수업코드 실습

```r
#install.packages(c("tidygraph",'ggraph'))
library(tidygraph)
library(ggraph)

feat <- read.csv('featuring.csv')
head(feat)
class(feat)
fg <- as_tbl_graph(feat)
class(fg)
fg
plot(fg)
ggraph(fg) +  geom_node_point()  +  geom_edge_link()
```
